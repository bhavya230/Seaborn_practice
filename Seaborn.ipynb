 # Correlation
 It is  how much 2 input features are related with each other .
 It is calculated for only the int/floating point features not for categorical features.

```python
df.corr(numeric_only=True) # shows correlation matrix
```

# Correlation Heatmap

```python
sns.heatmap(df.corr(numeric_only=True))
```

# JointPlots
It is a special type of visulization used to **analyze the relationship between two numerical variables** while simultaneously **showing their individual distributions**. 

## Scatter Jointplot

```python
sns.jointplot(x="total_bill",y="tip",data=df,color="red")
```

## Hexbin Jointplot

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind="hex")
```
