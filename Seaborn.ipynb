 # Correlation
 It is  how much 2 input features are related with each other .
 It is calculated for only the int/floating point features not for categorical features.

```python
df.corr(numeric_only=True) # shows correlation matrix
```

# Correlation Heatmap

```python
sns.heatmap(df.corr(numeric_only=True))
```

# JointPlots
It is a special type of visulization used to **analyze the relationship between two numerical variables** while simultaneously **showing their individual distributions**. 

## Scatter Jointplot

```python
sns.jointplot(x="total_bill",y="tip",data=df,color="red")
```

## Hexbin Jointplot

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind="hex")
```

## Regression Jointplot

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind="reg")
```

## KDE JOINTPLOT

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind="kde")
```

## Histogram Jointplot

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind="hist")
```

# Pairplots
 used to **visualize the relationship between mulitple variables at the same time**. It automatically generates a **matrix of plots showing how every variable relates to every other variable**

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

# Categotical Plots

# Count Plot
Countplot is used to **show the number of occurences(counts) of each category in a categorical variable**.

```python
sns.countplot(x="sex",data=df)
```

```python
sns.countplot(y="sex",data=df)
```

```python
sns.countplot(x="sex",hue="sex",data=df)
```

```python
sns.countplot(x="sex",hue="smoker",data=df)
```

## Barplot
A barplot is used to show the **relationship between a categorical variable and numerical variable**. Barplot shows a **statistical value(usually mean) of the numerical variable for each categpry.**

```python
df.head()
```

```python
sns.barplot(x="time",y="total_bill",data=df)
```

```python
sns.barplot(x="time",y="total_bill",hue="sex",data=df)
```

```python
sns.barplot(x="sex",y="total_bill",data=df)
```

```python
sns.barplot(x="sex",y="total_bill",estimator=np.median,data=df)
```
